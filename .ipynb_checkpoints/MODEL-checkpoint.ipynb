```python
import pandas as pd
import numpy as np
```

```python
niti = pd.read_csv('NITIDATA.csv')
```

```python
niti.head()
```

```python
niti.describe().T
```

```python
niti.corr()
```

# Split the Data

```python
from sklearn.model_selection import train_test_split
```

```python
X = niti.iloc[:, 0:-1]
```

```python
X.shape
```

```python
X.head()
```

```python
Y = niti.iloc[:, -1]
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y)
```

# Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

# Create the Model

```python
from sklearn.neural_network import MLPRegressor
```

```python
model = MLPRegressor(hidden_layer_sizes=(12, 12, 12, 12, 12, 12, 12, 12, 12, 12, 12), max_iter=5000)
```

```python
model.fit(X_train, Y_train)
```

```python
model.score(X_test, Y_test)
```

# Analyze the Model

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
Ya = Y_test
Yp = model.predict(X_test)
mse = mean_squared_error(Ya, Yp)
R2 = r2_score(Ya, Yp)
```

```python
'MSE : {}, RMSE : {}, R2-Score : {}'.format(mse, np.sqrt(mse), R2)
```

# Save the Model and the Scaler

```python
import pickle
from sklearn.externals import joblib
```

```python
filename = 'gdp_model.sav'
scalerfile = 'gdp_scale.sav'
pickle.dump(model, open(filename, 'wb'))
pickle.dump(scaler, open(scalerfile, 'wb'))

# using JobLib
joblib.dump(model, 'gdp.model')
joblib.dump(scaler, 'gdp.scaler')
```

# Load the Model and the Scaler

```python
loaded_model = pickle.load(open(filename, 'rb'))
loaded_scaler = pickle.load(open(scalerfile, 'rb'))
result = loaded_model.score(X_test, Y_test)
print(result)
```
